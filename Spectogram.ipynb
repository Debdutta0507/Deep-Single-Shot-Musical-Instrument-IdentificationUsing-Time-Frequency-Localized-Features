```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import glob
import os
inputfolder='/content/drive/My Drive/project_images/Data_set/to_be_uploaded/ClarinetG5'
i=0
for sound in glob.glob(inputfolder+"/*.wav"):
  i=i+1
  #try:
  sample_rate, samples = wavfile.read(sound)
  frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate,window=('hann'))
  
 
  plt.pcolormesh(times,frequencies,np.log10(spectrogram))



  plt.ylabel('Frequency [Hz]')
  plt.xlabel('Time [sec]')
 
 

  
  plt.imshow(np.log10(spectrogram))
  


 
   #plt.gca().axes.get_yaxis().set_visible(False)
   #plt.gca().axes.get_xaxis().set_visible(False)
  
  plt.savefig("/content/drive/My Drive/project_images/Data_set/to_be_uploaded/Clarinet_G5/"+"Clarinet"+str(i))
  #plt.imshow(np.log10(spectrogram))
  ### continue
```
